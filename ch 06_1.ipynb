```python
import numpy as np
import matplotlib.pyplot as plt
```

넘파이 와 맷플롯립 패키지 임포트

```python
fruits = np.load('fruits_300.npy')
```

fruits_300.npy 파일에 저장된 NumPy 배열을 불러와 fruits 변수에 저장

```python
print(fruits.shape)
```

fruits라는 NumPy 배열의 차원(shape) 을 출력
300개의 이미지(또는 샘플)

각 이미지 크기: 100 × 100 픽셀

```python
print (fruits[0,0,:])
```

첫 번째 행에 있는 픽셀 100개에 들어 있는 값을 출력
흑백 사진을 담고
있으므로 0〜255까지의 정숫값 가짐

```python
plt.imshow(fruits[0], cmap='gray')
plt.show()
```

fruits 배열에서 첫 번째 이미지(0번 인덱스) 를 선택해 흑백으로 화면에 표시

```python
plt.imshow(fruits[0], cmap='gray_r')
plt.show()
```

일한 이미지지만 밝기 표현이 반전된 흑백 이미지로 표시
gray → 밝은 값이 밝게, 어두운 값이 어둡게 표현되는 일반 흑백

gray_r → _r가 붙으면 reversed(반전) 된 colormap

```python
fig, axs = plt.subplots(1,2)
axs[0].imshow(fruits[100], cmap='gray_r')
axs[1].imshow(fruits[200], cmap='gray_r')
plt.show()
```

fruits 데이터셋의 100번과 200번 이미지를 흑백 반전(gray_r) 으로
좌우 나란히 보여줌

```python
apple = fruits[0:100].reshape(-1, 100*100)
pineapple = fruits[100:200].reshape(-1, 100*100)
banana = fruits[200:300].reshape(-1, 100*100)
```

fruits[0:100]

fruits 배열에서

0번 ~ 99번까지 100장의 이미지를 선택

보통 shape은 (100, 100, 100) (100장 × 100×100 픽셀)

 .reshape(-1, 100*100)

각 이미지(100×100)를
1차원 벡터(10000픽셀) 로 펼쳐서
(100, 10000) 형태로 바꿈

-1은 “행 수는 자동으로 맞춰라”는 의미이므로
100장 → 자동으로 100

```python
print(apple.shape)
```

apple 배열의 크기(shape) 를 출력

apple 배열에는 100장의 이미지가 있고

각 이미지는 100×100 = 10000 픽셀로 펼쳐짐


```python
print(apple.mean(axis=1))
```

apple 배열의 각 이미지별 평균 픽셀 값을 계산하고 출력


```python
plt.hist(apple.mean(axis=1), alpha=0.8, label='apple')
plt.hist(pineapple.mean(axis=1), alpha=0.8, label='pineapple')
plt.hist(banana.mean(axis=1), alpha=0.8, label='banana')
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(1,3, figsize=(20,5))
axs[0].bar(range(10000), apple.mean(axis=0))
axs[1].bar(range(10000), pineapple.mean(axis=0))
axs[2].bar(range(10000), banana.mean(axis=0))
plt.show()
```

각 과일 클래스별 평균 이미지의 픽셀 값을 1차원 바 차트로 나타냄
apple.mean(axis=0) → 100장의 사과 이미지에서 각 픽셀 위치의 평균값 계산

range(10000) → 10,000개 픽셀 위치를 x축으로 사용

bar → 각 픽셀 평균값을 높이로 하는 막대그래프 생성

1행 3열 subplot → 사과, 파인애플, 바나나를 나란히 비교

```python
apple_mean = apple.mean(axis=0).reshape(100, 100)
pineapple_mean = pineapple.mean(axis=0).reshape(100,100)
banana_mean = banana.mean(axis=0).reshape(100, 100)
fig, axs = plt.subplots(1, 3, figsize=(20,5))
axs[0].imshow(apple_mean, cmap='gray_r')
axs[1].imshow(pineapple_mean, cmap='gray_r')
axs[2].imshow(banana_mean, cmap='gray_r')
plt.show
```

100장씩의 이미지 데이터를 평균 내어 각 과일의 대표 이미지를 만들고,
이를 2D 흑백 이미지로 시각화하여 세 과일을 나란히 비교

apple.mean(axis=0).reshape(100,100):

100장의 사과 이미지를 픽셀별 평균 계산 → 길이 10,000 벡터

reshape(100,100): 다시 2D 이미지 형태로 변환 즉, 사과 평균 이미지 생성

pineapple_mean, banana_mean:

각각 파인애플, 바나나 평균 이미지 생성

fig, axs = plt.subplots(1, 3, figsize=(20,5)):

1행 3열 subplot 생성 → 세 과일 이미지를 나란히 비교

axs[i].imshow(..., cmap='gray_r'):

각 subplot에 평균 이미지 표시

gray_r → 흑백 반전(colormap 반전)

plt.show()

화면에 세 과일 평균 이미지 출력

```python
abs_diff = np.abs(fruits - apple_mean)
abs_mean = np.mean(abs_diff, axis=(1,2))
print(abs_mean.shape)
```

모든 이미지가 사과 평균 이미지와 얼마나 다른지를 계산하는 과정, 값이 작을수록 사과 평균 이미지와 유사, 값이 클수록 차이가 크다는 뜻

abs_diff는 shape (300, 100, 100):
각 이미지가 사과 평균 이미지와 얼마나 다른지 픽셀 단위로 절대값 차이를 나타냄

abs_mean = np.mean(abs_diff, axis=(1,2)):
이미지 내부의 모든 픽셀 평균
각 이미지마다 사과 평균 이미지와의 평균 절대 차이를 하나의 숫자로 표현

출력결과: 300장의 이미지 각각에 대해, 사과 평균 이미지와의 평균 절대 차이를 계산한 값

```python
apple_index = np.argsort(abs_mean)[:100]
apple_index = apple_index.reshape(10,10)
fig, axs = plt.subplots(10,10, figsize=(10,10))
for i in range(10):
  for j in range(10):
    axs[i, j].imshow(fruits[apple_index[i, j]], cmap='gray_r')
    axs[i, j].axis('off')
  plt.show()
```

apple_index = np.argsort(abs_mean)[:100]:
사과 평균 이미지와 가장 비슷한 100장의 이미지 인덱스를 구한 것

apple_index = apple_index.reshape(10,10):
100개의 인덱스를 10×10 배열로 변환

사과 평균 이미지와 가장 유사한 100장의 사과 이미지를 10×10 그리드로 시각화 하는 코드, 이미지가 중심에 가까울수록 평균 이미지와 더 유사
