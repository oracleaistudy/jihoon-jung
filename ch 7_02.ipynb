```python
import keras
(train_input, train_target), (test_input, test_target) = \
keras.datasets.fashion_mnist.load_data()
```

```python
from sklearn.model_selection import train_test_split

train_scaled = train_input / 255.0
train_scaled = train_scaled.reshape(-1, 28*28)
train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42
)
```

train_scaled = train_input / 255.0:

픽셀값(0~255)을 0~1 범위로 정규화

신경망 학습이 더 잘 되도록 하는 기본 전처리

train_scaled = train_scaled.reshape(-1, 28*28):

28×28 이미지를 **1차원 벡터(784)**로 펼침

신경망(Dense layer)이 1차원 입력을 받아야 하기 때문

-1은 “자동으로 개수 맞춰라”라는 뜻

전체 훈련 데이터 중 **20%를 검증 데이터(val)**로 분리

80%는 train, 20%는 validation

random_state=42 → 재실행해도 동일하게 섞이도록 고정

```python
inputs = keras.layers.Input(shape=(784,))
dense1 = keras. layers.Dense(100, activation='sigmoid')
dense2 = keras.layers.Dense(10, activation='softmax')
```

입력층 정의

28×28 이미지를 펼친 784차원 벡터가 들어옴

첫 번째 은닉층(Dense layer)

뉴런 100개

활성화 함수: sigmoid

출력층

클래스 10개(Fashion-MNIST)

다중분류용 softmax

```python
model = keras.Sequential([inputs, dense1, dense2])
```

inputs : 입력층 (shape = 784)

dense1 : 은닉층 100개, sigmoid

dense2 : 출력층 10개, softmax

이들을 순서대로 쌓아 하나의 모델을 만듦

```python
model.summary()
```

층 이름

출력 크기(Output shape)

파라미터 수(Weights 개수)

전체 파라미터 총합 나타냄

```python
model = keras.Sequential([
    keras.layers.Input(shape=(784,)),
    keras.layers.Dense(100, activation='sigmoid', name='은닉층'),
    keras.layers.Dense(10, activation='softmax', name='출력층')
], name='패션 MNIST 모델')
```

아래 코드는 Keras Sequential API로 간단한 신경망(MLP)을 정의한 것입니다.
패션 MNIST(Fashion MNIST) 데이터를 분류하기 위한 모델 구조

```python
model.summary()
```

```python
model = keras.Sequential()
model.add(keras.layers.Input(shape=(784,)))
model.add(keras.layers.Dense(100, activation='sigmoid'))
model.add(keras.layers.Dense(10, activation='softmax'))
```

784개의 입력을 받아
은닉층 100개(sigmoid),
출력층 10개(softmax)를 가진 신경망을 만든 코드

즉, 패션 MNIST 같은 10개 클래스 분류를 위한 기본적인 MLP 모델

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

sparse categorical crossentropy를 손실로 사용하고 정확도를 평가 지표로 삼아
train_scaled와 train_target을 이용해서 모델을 5번(epoch) 동안 학습

```python
model= keras.Sequential()
model.add(keras.layers.Input(shape=(28,28)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
```

28×28 이미지를 받아 10개 클래스 중 하나로 분류하는 MLP 모델

```python
model.summary()
```

```python
(train_input, train_target), (test_input, test_target) =\
keras.datasets.fashion_mnist.load_data()
train_scaled = train_input / 255.0
train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

패션 MNIST 데이터를 0~1로 정규화하고, 학습/검증용으로 나눔

```python
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

습 데이터를 이용해 모델을 5회 반복 학습시키고, 손실과 정확도를 계산하도록 설정

```python
model.evaluate(val_scaled, val_target)
```

검증 데이터(val_scaled, val_target)를 사용해 모델 성능 평가

학습이 끝난 모델이 얼마나 잘 맞추는지 확인하는 단계

```python
model = keras.Sequential()
model.add(keras.layers.Input(shape=(28,28)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
```

28×28 이미지를 입력으로 받는 순차 신경망을 만듬
입력 이미지를 Flatten 층으로 784차원 1차원 벡터로 변환
은닉층에는 100개의 뉴런과 ReLU 활성화 함수를 사용
출력층은 10개 뉴런과 Softmax로 각 클래스에 속할 확률을 출력

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

Adam 옵티마이저와 sparse categorical crossentropy 손실 함수로 모델 학습을 준비


학습 중에는 **정확도(accuracy)를 평가 지표(metric)**로 사용

train_scaled와 train_target 데이터를 이용해 모델을 학습시킴


전체 학습 데이터를 5번(epoch) 반복하여 모델이 점점 성능을 개선


```python
model.evaluate(val_scaled, val_target)
```

학습된 모델을 검증 데이터로 평가해 손실과 정확도를 확인
